```python
import turtle
t=turtle.Turtle()
t.speed(10)
a=int(input(" dy nt v: "))
b=int(input("bn mun c my ngi nh: "))
turtle.bgcolor("sky blue")
t.speed(10)
#turtle.bgcolor(blue)
#v mt tri
t.penup()
t.goto(300,300)
t.pendown()
t.color("red","yellow")
t.begin_fill()
t.pensize(5)
for i in range(36):
    t.fd(30)
    t.rt(150)
    t.fd(30)
    t.rt(30)
    t.rt(170)
t.end_fill()
#v ci cy
t.penup()
t.goto(-350,-175)
t.pendown()
t.pensize(50)
t.pencolor("brown")
t.lt(90)
t.fd(100)
t.pensize(5)
t.penup()
t.rt(90)

t.pendown()
t.color("green","green")
t.begin_fill()
t.circle(100)
t.end_fill()
#ta  ngi nh th 1
t.penup()
t.goto(-200,0)
t.pendown()
for i in range(b):    
    
    t.color("black","yellow")
    t.begin_fill()
    t.penup()
    t.pendown()
    t.pensize(a) 
    for i in range(2):
        t.fd(180)
        t.rt(90)
        t.fd(200)
        t.rt(90)
    t.end_fill()
    # nc nh
    t.color("black","red")
    t.begin_fill()
    t.lt(60)
    t.fd(180)
    t.rt(120)
    t.fd(180)
    t.lt(150)
    t.end_fill()
    #v ng khi
    t.fd(100)
    t.lt(90)
    t.fd(50)
    t.lt(90)
    t.fd(15)
    t.penup()
    t.fd(285)
    t.pendown()


    #v cnh ca
    t.color("black","brown")
    t.begin_fill()
    t.rt(180)
    t.fd(100)
    t.lt(90)
    t.fd(80)
    t.lt(90)
    t.fd(100)
    t.end_fill() 

    #v ca s

    t.penup()
    #t.goto(30,-30)
    t.lt(180)
    t.fd(120)
    t.pendown()
    t.color("black","white")
    t.begin_fill()
    for i in range(4):
        t.fd(40)
        t.lt(90)
    t.end_fill()  
    t.penup()
    t.rt(90)
    t.fd(80)
    t.pendown()  
    t.color("black","white")
    t.begin_fill()
    for i in range(4):
        t.fd(40)
        t.lt(90)
    t.end_fill()

    t.penup()
    t.fd(50)
    t.lt(90)
    t.fd(80)
    t.rt(90)
    t.pendown()

    
    
turtle.mainloop()    
```
